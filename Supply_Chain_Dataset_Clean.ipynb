#Data Cleaning Process

This is the python code to check null-values or duplicated in the 'Supply Chain Dataset'.

The dataset can be downloaded [here](https://docs.google.com/spreadsheets/d/1po7nlVs59W-nxm8coIKeOUsWYZI4h6X_xD_vR_uEcOI/edit?usp=sharing).

```python
import pandas as pd
```

##Import Data

```python
df = pd.read_csv("Supply Chain Dataset - DataCoSupplyChainDataset.csv")
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df.info()
```

##Checking Null Values

```python
df.isnull().sum()
```

Because `Product Description` and `Order Zipcode` columns have more than 75% null-values, it is recommended to drop the columns.

```python
df = df.drop(['Order Zipcode', 'Product Description'], axis=1)
```

```python
df.info()
```

There are 2 columns with some missing values, they are: `Customer Lname` and `Customer Zipcode`.

###Fill The Null-Values

```python
df[df['Customer Zipcode'].isnull()]
```

As seen on the table above, there seems to be some mismatch values, as an example: `Customer State` but it is an integer instead of string.

Let see if it is true.

Try to find the `Customer Zipcode` with the value of `Customer State` from the table above (95758).

```python
df[df['Customer Zipcode'] == 95758]
```

Based on the result, not only for `Customer Zipcode`, but `Customer State`, `Customer City`, `Customer Street` are all in wrong places.

Change all those columns into the correct values, except for `Customer Street` because there is no reference and it is still logical to be in the same city as `Customer City`.

There is also no other data for these customers.

```python
df[(df['Customer Lname'] == 'Elliott') & (df['Customer Fname'] == 'Zena')]
```

Replacing the values.

```python
df.loc[35704, ['Customer Zipcode', 'Customer State', 'Customer City']] = [95758, 'CA', 'Elk Grove']
df.loc[46440, ['Customer Zipcode', 'Customer State', 'Customer City']] = [95758, 'CA','Elk Grove']
df.iloc[[35704, 46440]]
```

```python
df[df['Customer Zipcode'] == 91732]
```

```python
df.loc[82511, ['Customer Zipcode', 'Customer State', 'Customer City']] = [91732, 'CA', 'El Monte']
df.iloc[[82511]]
```

Check the null-values in `Customer Lname` column.

```python
df[df['Customer Lname'].isnull()]
```

```python
df["Customer Lname"].fillna(df["Customer Fname"], inplace = True)
df[df['Customer Lname'].isnull()]
```

Re-check a column.

```python
df.iloc[[41422]]
```

```python
df.info()
```

There is no null-values.

##Checking Duplicates

```python
df.duplicated().sum()
```

There is no duplicated values.

##Saving the Dataframe

Saving the dataframe into csv files to be visualized using Tableau.

```python
df.to_csv('Supply_Chain_Dataset_Clean.csv', sep=',', index=False, encoding='utf-8')
```
